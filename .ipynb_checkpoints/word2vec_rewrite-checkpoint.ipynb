```python
import matplotlib.pyplot as plt
import numpy as np

import re  # For preprocessing
import pandas as pd  # For data handling
from time import time  # To time our operations
from collections import defaultdict  # For word frequency

plt.plot([1,2,3])
```

```python
import spacy  # For preprocessing

import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(42)
```

### Load text

```python
#load text
import io
f = io.open("20140615-wiki-en_000000.txt", mode="r", encoding="utf-8")
text_raw = f.read()[10000:510000]
```

### Split into sntnc

```python
from nltk import tokenize
text_tokenized = tokenize.sent_tokenize(text_raw)
```

### Cleaning via spacy

```python
#Removes non-alphabetic characters:
brief_cleaning = (re.sub("[^A-Za-z']+", ' ', str(sent)).lower() for sent in text_tokenized)

texts = brief_cleaning

text_clean = []

nlp = spacy.load("en_core_web_sm")
for doc in nlp.pipe(texts, disable=['ner', 'parser']):
    # Do something with the doc here
    for token in doc:
        if not token.is_stop:
            if token.lemma_ not in ['s', 're']:
                text_clean.append(token.lemma_)
```

```python
#create vocab

print(text_clean[100:120])

vocab = set(text_clean)
vocab_size = len(vocab)
print('text_size:', len(text_clean))
print('vocab_size: ', vocab_size)


w2i = {w:i for i, w in enumerate(vocab)}
i2w = {i:w for i, w in enumerate(vocab)}
```

```python
boy_count = 0
girl_count = 0
man_count = 0
woman_count = 0
for i in text_clean:
    if i == 'boy':
        boy_count += 1
    elif i == 'man':
        man_count += 1
    elif i == 'girl':
        girl_count += 1
    elif i == 'woman':
        woman_count += 1
        
print(boy_count)
print(girl_count)
print(man_count)
print(woman_count)
```

### cbow dataset

```python
CONTEXT_SIZE = 2

def create_cbow_dataset(text):
    data = []
    for i in range(CONTEXT_SIZE, len(text)-CONTEXT_SIZE):
        context = [text[i-k] for k in range(1,CONTEXT_SIZE+1)] + [text[i+k] for k in range(1,CONTEXT_SIZE+1)]
        target = text[i]
        
        data.append((context, target))
    return data
```

```python
cbow_train = create_cbow_dataset(text_clean)
cbow_train[2]
```

```python
class CBOW(nn.Module):
    def __init__(self, vocab_size, embd_size, context_size, hidden_size):
        super(CBOW, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embd_size)
        self.linear1 = nn.Linear(2*context_size*embd_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, vocab_size)
    
    def forward(self, inputs):
        stacked_embd = self.embeddings(inputs).view((1, -1))
        lin1 = self.linear1(stacked_embd)
        hid = F.relu(lin1)
        lin2 = self.linear2(hid)
        log_probs = F.log_softmax(lin2)
        return log_probs
```

```python
embd_size = 50
learning_rate = 0.01
n_epochs = 10

losses = []

def train_cbow(losses):
    hidden_size = 512
    loss_fn = nn.NLLLoss()
    model = CBOW(vocab_size, embd_size, CONTEXT_SIZE, hidden_size)
    print(model)
    optimizer = optim.SGD(model.parameters(), lr = learning_rate)
    
    for epoch in range(n_epochs):
        print(f'epoch num {epoch}', end =' ')
        total_loss = 0.0
        for context, target in cbow_train:
            context_idx = [w2i[w] for w in context]
            context_var = Variable(torch.tensor(context_idx))
            
            model.zero_grad()
            log_probs = model(context_var)
            
            loss = loss_fn(log_probs, Variable(torch.tensor([w2i[target]])))
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        losses.append(total_loss)
    
    return model, losses
```

```python
model_trained, losses = train_cbow(losses)
```

```python
plt.plot(losses)
plt.yscale('log')
```
